```
(*Andrews' rational T(r,h) functions*)
Option[T, MemoryConstraint -> Infinity];
T[r_, h_] := Module[{x},
    x = 0;
    If[h == 0 || h == 1,
        (h - 1)^2,
        Do[
            x = x - qbin[q, k, r + 2*k] * T[r + 2*k, h - 2*k],
            {k, 1, Floor[h/2]}
        ];
        Normal[Series[x,{q,0,1000}]]
    ]
];
```
